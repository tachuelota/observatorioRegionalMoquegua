### **REGIÓN MOQUEGUA REPORTE DESCRIPTIVO DE VICTIMIZACIÓN **

1.	Evolución de la población víctima de algún hecho delictivo, 2021 – 2023

```python
# @title
import matplotlib.pyplot as plt

# Datos proporcionados
years = [2021, 2022, 2023]
numerador = [14629, 19428, 22278]
denominador = [122158, 125355, 130538]
porcentaje = [12.0, 15.5, 17.1]

# Crear la figura y los ejes
fig, ax1 = plt.subplots()

# Graficar el porcentaje en un eje
ax1.plot(years, porcentaje, color='tab:blue', marker='o', label='Porcentaje', linestyle='-')
ax1.set_xlabel('Año')
ax1.set_ylabel('Porcentaje de víctimas', color='tab:blue')
ax1.tick_params('y', colors='tab:blue')
ax1.set_ylim(0, max(porcentaje) * 1.1)

# Crear un segundo eje y graficar numerador y denominador
ax2 = ax1.twinx()
ax2.bar(years, numerador, color='tab:orange', alpha=0.5, label='Numerador')
ax2.plot(years, denominador, color='tab:green', marker='o', label='Denominador', linestyle='--')
ax2.set_ylabel('Cantidad de víctimas', color='tab:orange')
ax2.tick_params('y', colors='tab:orange')
ax2.set_ylim(0, max(max(numerador), max(denominador)) * 1.1)

# Añadir leyendas y título
#fig.tight_layout()
fig.legend()
plt.title('Evolución de la población víctima de algún hecho delictivo en Moquegua (2021-2023)')

# Mostrar la gráfica
plt.grid(True)
plt.show()
```

1.1.	Porcentaje de victimización por el delito de robo e intento de robo de dinero, cartera y celular

```python
# @title
import matplotlib.pyplot as plt

# Datos
años = [2021, 2022, 2023]
porcentaje_victimizacion = [4.5, 7.3, 7.0]

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(años, porcentaje_victimizacion, marker='o', color='b', linestyle='-')

# Etiquetas y título
plt.title('Porcentaje de victimización por robo e intento de robo')
plt.xlabel('Año')
plt.ylabel('Porcentaje')
plt.xticks(años)

# Mostrar porcentaje en cada punto
for i, porcentaje in enumerate(porcentaje_victimizacion):
    plt.text(años[i], porcentaje, f'{porcentaje}%', ha='right', va='bottom')

# Mostrar la cuadrícula
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

3.	Evolución de la población víctima de algún hecho delictivo que no denunció el hecho, 2021 – 2023

```python
# @title
import matplotlib.pyplot as plt
import numpy as np

# Datos
años = [2021, 2022, 2023]
denunciaron_porcentaje = [20.4, 12.8, 15.1]
no_denunciaron_porcentaje = [79.6, 87.2, 84.9]

# Graficar
fig, ax = plt.subplots(figsize=(10, 6))

# Barra para "Denunciaron"
bar1 = ax.bar(años, denunciaron_porcentaje, color='b', label='Denunciaron')

# Barra para "No denunciaron"
bar2 = ax.bar(años, no_denunciaron_porcentaje, bottom=denunciaron_porcentaje, color='r', label='No denunciaron')

# Añadir etiquetas y título
ax.set_xlabel('Año')
ax.set_ylabel('Porcentaje')
ax.set_title('Evolución de la población víctima de algún hecho delictivo')
ax.set_xticks(años)
ax.legend()

# Mostrar porcentaje en cada barra
for bars in [bar1, bar2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height}%', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords='offset points', ha='center', va='bottom')

# Mostrar la gráfica
plt.grid(True, axis='y')
plt.show()
```
